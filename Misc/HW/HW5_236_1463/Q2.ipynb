```

def getOrthogonalDistance(p0,p1,p):
  return np.abs(np.cross(p1-p0,p-p0)/np.linalg.norm(p1-p0))

def getIntersectionPoint(p0,p1,p):

    x1=p0[0]
    y1=p0[1]
    x2=p1[0]
    y2=p1[1]
    x3=p[0]
    y3=p[1]
    px = x2-x1
    py = y2-y1
    dAB = px*px + py*py
    u = ((x3 - x1) * px + (y3 - y1) * py) / dAB
    x = x1 + u * px
    y = y1 + u * py
    return x,y
```

```
p0=np.array([40,20])
p1=np.array([80,20])
p=np.array([40,40])
dist=getOrthogonalDistance(p0,p1,p)
q=getIntersectionPoint(p0,p1,p)
q
```

```
import matplotlib.pyplot as plt
x1 = [p0[0], p1[0]]
y1=  [p0[1], p1[1]]

x2 = [p[0], q[0]]
y2=  [p[1], q[1]]

plt.plot(x1, y1, marker = 'o')
plt.plot(x2, y2, marker = 'o')
plt.show()
```

Projection of a on b (a1), and rejection of a from b (a2).

Projection_and_rejection.png

```
def projection(u, v):
    proj_of_u_on_v = v * np.dot(u, v) / np.dot(v, v)
    return  proj_of_u_on_v

def rejection(u, v):
    return u - projection(u, v)

def getOrthogonalVector(p0,p1,p):
  u=p-p0
  v=p1-p0
  return rejection(u, v)

r=getOrthogonalVector(p0,p1,p)

p-r
```
